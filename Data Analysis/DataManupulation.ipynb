```python
## Handling missing values 

df.isnull().any()
```

```python
df_filled = df.fillna(0)
df_filled
```

```python
## filling missing valueswith mean of the column 

df['Salary_fillNA'] = df['Salary'].fillna(df['Salary'].mean())
df
```

```python
df.dtypes
```

```python
## Renaming columns 

df.rename(columns={'Join_Date':'Sales-Date'})
```

```python
## Chnage the datatypes

df['value_new'] = df['Age'].fillna(df['Age'].mean()).astype(int)

df.rename(columns={'value_new':'Age_new'})
```

```python
df['New Value'] = df['Age'].apply(lambda x:x*2)
df.head()
```

```python
## data Aggregating and Grouping

df.head()
```

```python
grouped_mean=df.groupby('Name')['value_new'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Department','Name'])['value_new'].sum()
print(grouped_sum)
```

```python
## Aggrefgate multiple functions

grouped_agg = df.groupby('Name')['value_new'].agg(['mean','sum','count'])
print(grouped_agg)
```

```python
## Merginf and joing dataFrames

##Create sample DataFrames
df1=pd.DataFrame({'key':['A','B','c'], 'value1':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','d'], 'value2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
## Merge the DataFrames on the 'Key Columns'"Online Sales Data.csv"
pd.merge(df1,df2,on="key",how="inner")
```

```python
## Merge the DataFrames on the 'Key Columns'"Online Sales Data.csv"
pd.merge(df1,df2,on="key",how="outer")
```

```python
## Merge the DataFrames on the 'Key Columns'"Online Sales Data.csv"
pd.merge(df1,df2,on="key",how="left")
```

```python
## Merge the DataFrames on the 'Key Columns'"Online Sales Data.csv"
pd.merge(df1,df2,on="key",how="right")
```
